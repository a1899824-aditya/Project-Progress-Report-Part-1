##### Correcting The Skewness

```python
skewness = df.skew()
print(skewness)
```

As we can see from the above information related to skewness we can Particulate, ParticulateMax, ParticulateMin, RainIntensity, RainIntensityMax, RainIntensityMin, WindSpeedMax, and WindSpeedMin have high skewness . These features indicate that most of their respective values are clustered on left hand side applying a transformation will help to normalize the distribution and hence improving the model performance . On the other hand feature RelativeHumidity  show negative skewness . Since we haven't done feature selection related process , we will include this feature and hence transform RelativeHumidity feature as well. 

```python
import numpy as np 
df['Particulate'] = np.log1p(df['Particulate'])
df['ParticulateMax'] = np.log1p(df['ParticulateMax'])
df['ParticulateMin'] = np.log1p(df['ParticulateMin'])

df['RainIntensity'] = np.log1p(df['RainIntensity'])
df['RainIntensityMax'] = np.log1p(df['RainIntensityMax'])
df['RainIntensityMin'] = np.log1p(df['RainIntensityMin'])

df['WindSpeedMax'] = np.log1p(df['WindSpeedMax'])
df['WindSpeedMin'] = np.log1p(df['WindSpeedMin'])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd

# Initialize a dictionary to store RMSE and R-squared values for each feature
feature_metrics = {}

# Iterate through all features
for feature in X_train_rfl.columns:
    print(f"Processing feature: {feature}")
    
    # Train on all features except the current one
    temp_X_train = X_train_rfl.drop(columns=[feature])
    temp_X_test = X_test_rfl.drop(columns=[feature])
    
    # Train a RandomForestRegressor
    model = RandomForestRegressor(n_estimators=50, random_state=42, oob_score=True, bootstrap=True, n_jobs=-1)
    model.fit(temp_X_train, y_train_rfl.values.ravel())
    
    # Make predictions and evaluate metrics
    y_pred = model.predict(temp_X_test)
    rmse = np.sqrt(mean_squared_error(y_test_rfl, y_pred))
    r2 = model.score(temp_X_test, y_test_rfl)
    
    # Store metrics for the current feature
    feature_metrics[feature] = {
        'OOB_Score': model.oob_score_,
        'RMSE': rmse,
        'R2': r2
    }

# Convert feature metrics into a DataFrame for better visualization
feature_metrics_df = pd.DataFrame(feature_metrics).T.sort_values(by='RMSE', ascending=True)

# Display the feature metrics
print("\nFeature Metrics (Sorted by RMSE):")
print(feature_metrics_df)


feature_metrics_df.to_csv("feature_metrics_rfl.csv", index=True)

# Output the features with the best performance
print("\nFeatures with the Best RMSE and R2 Scores:")
print(feature_metrics_df.head())
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.utils import resample
import pandas as pd
import numpy as np

# Step 1
desired_samples = 500
balanced_df = pd.DataFrame(columns=df.columns)

for synop_code in df['SYNOPCode'].unique():
    synop_data = df[df['SYNOPCode'] == synop_code]
    if len(synop_data) > desired_samples:
        synop_sampled = synop_data.sample(n=desired_samples, random_state=42)
    else:
        synop_sampled = resample(synop_data, replace=True, n_samples=desired_samples, random_state=42)
    balanced_df = pd.concat([balanced_df, synop_sampled])

balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Print the updated SYNOP code distribution
print("Updated Synop Code Distribution:\n", balanced_df['SYNOPCode'].value_counts())

# Step 2: Define Features and Target Variables
X = balanced_df.drop(columns=['SYNOPCode', 'FSO_Att', 'RFL_Att'])
y_fso = balanced_df['FSO_Att']
y_rfl = balanced_df['RFL_Att']

# Split into train-test sets for both FSO_Att and RFL_Att
X_train_fso, X_test_fso, y_train_fso, y_test_fso = train_test_split(X, y_fso, test_size=0.2, random_state=42)
X_train_rfl, X_test_rfl, y_train_rfl, y_test_rfl = train_test_split(X, y_rfl, test_size=0.2, random_state=42)

# Step 3: Define Backward Elimination Function
def backward_elimination(X_train, X_test, y_train, y_test, target_name):
    print(f"\n--- Performing Backward Elimination for {target_name} ---")
    remaining_features = list(X_train.columns)
    best_oob_score = 0
    best_features = remaining_features.copy()
    metrics = {}

    while len(remaining_features) > 1:
        print(f"Remaining features: {remaining_features}")
        model = RandomForestRegressor(n_estimators=100, random_state=42, oob_score=True, bootstrap=True)
        model.fit(X_train[remaining_features], y_train)
        
        oob_score = model.oob_score_
        y_pred = model.predict(X_test[remaining_features])
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        r2 = r2_score(y_test, y_pred)

        print(f"OOB Score: {oob_score}, RMSE: {rmse}, R2: {r2}")
        
        # Store metrics
        metrics = {'OOB_Score': oob_score, 'RMSE': rmse, 'R2': r2}
        
        # Update best features if the new model is better
        if oob_score > best_oob_score:
            best_oob_score = oob_score
            best_features = remaining_features.copy()
        
        # Identify the least important feature and remove it
        feature_importances = model.feature_importances_
        least_important_feature = remaining_features[np.argmin(feature_importances)]
        print(f"Removing least important feature: {least_important_feature}")
        remaining_features.remove(least_important_feature)

    print(f"\nFinal Metrics for {target_name}: {metrics}")
    print(f"Selected Features for {target_name}: {best_features}")
    return best_features, metrics

# Step 4: Perform Backward Elimination for FSO_Att
print("\n--- Processing FSO_Att ---")
fso_features, fso_metrics = backward_elimination(X_train_fso, X_test_fso, y_train_fso, y_test_fso, "FSO_Att")

# Step 5: Perform Backward Elimination for RFL_Att
print("\n--- Processing RFL_Att ---")
rfl_features, rfl_metrics = backward_elimination(X_train_rfl, X_test_rfl, y_train_rfl, y_test_rfl, "RFL_Att")

# Print Final Metrics
print("\n--- Final Metrics ---")
print(f"FSO_Att Metrics: {fso_metrics}")
print(f"RFL_Att Metrics: {rfl_metrics}")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.utils import resample
import numpy as np
import pandas as pd

# Step 1: Downsample the dataset
desired_samples = 500  # Choose an appropriate sample size per SYNOP code
print("Downsampling the dataset...")

# Create an empty DataFrame to store downsampled data
balanced_df = pd.DataFrame(columns=df.columns)

# Perform downsampling for each SYNOP code
for synop_code in df['SYNOPCode'].unique():
    synop_data = df[df['SYNOPCode'] == synop_code]
    if len(synop_data) > desired_samples:
        synop_sampled = synop_data.sample(n=desired_samples, random_state=42)
    else:
        synop_sampled = resample(synop_data, replace=True, n_samples=desired_samples, random_state=42)
    balanced_df = pd.concat([balanced_df, synop_sampled])

balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Print the updated SYNOP code distribution
print("Updated Synop Code Distribution:\n", balanced_df['SYNOPCode'].value_counts())

# Step 2: Split into features and target variables
X = balanced_df.drop(columns=['SYNOPCode', 'FSO_Att', 'RFL_Att'])
y_fso = balanced_df['FSO_Att']
y_rfl = balanced_df['RFL_Att']
synop_codes = balanced_df['SYNOPCode'].unique()

# Step 3: Define Hyperparameter Tuning Function
def tune_hyperparameters(X_train, y_train, target_name):
    print(f"Hyperparameter tuning for {target_name}...")
    param_grid = {
        'n_estimators': [100, 200],
        'max_depth': [10, 20, None],
        'min_samples_split': [2, 5],
        'min_samples_leaf': [1, 2],
        'max_features': ['sqrt']
    }

    grid_search = GridSearchCV(
        estimator=RandomForestRegressor(random_state=42, oob_score=True, bootstrap=True),
        param_grid=param_grid,
        cv=3,
        scoring='neg_mean_squared_error',
        verbose=2,
        n_jobs=-1
    )
    grid_search.fit(X_train, y_train)
    print(f"Best Parameters for {target_name}:", grid_search.best_params_)
    return grid_search.best_params_

# Split train-test for the entire dataset
X_train, X_test, y_train_fso, y_test_fso = train_test_split(X, y_fso, test_size=0.2, random_state=42)
_, _, y_train_rfl, y_test_rfl = train_test_split(X, y_rfl, test_size=0.2, random_state=42)

# Tune hyperparameters for RFL_Att and FSO_Att
best_params_rfl = tune_hyperparameters(X_train, y_train_rfl, "RFL_Att")
best_params_fso = tune_hyperparameters(X_train, y_train_fso, "FSO_Att")

# Step 4: Train Models for Each SYNOP Code
metrics = {}

for synop_code in synop_codes:
    print(f"\nProcessing SYNOP Code: {synop_code}")

    # Filter data for the current SYNOP code
    synop_data = balanced_df[balanced_df['SYNOPCode'] == synop_code]
    X_synop = synop_data.drop(columns=['SYNOPCode', 'FSO_Att', 'RFL_Att'])
    y_fso_synop = synop_data['FSO_Att']
    y_rfl_synop = synop_data['RFL_Att']

    # Split train-test for the current SYNOP code
    X_train_synop, X_test_synop, y_train_fso_synop, y_test_fso_synop = train_test_split(
        X_synop, y_fso_synop, test_size=0.2, random_state=42
    )
    _, _, y_train_rfl_synop, y_test_rfl_synop = train_test_split(
        X_synop, y_rfl_synop, test_size=0.2, random_state=42
    )

    # Train FSO_Att model
    fso_model = RandomForestRegressor(**best_params_fso, random_state=42, oob_score=True, bootstrap=True)
    fso_model.fit(X_train_synop, y_train_fso_synop)
    y_pred_fso = fso_model.predict(X_test_synop)
    fso_rmse = np.sqrt(mean_squared_error(y_test_fso_synop, y_pred_fso))
    fso_r2 = r2_score(y_test_fso_synop, y_pred_fso)

    # Train RFL_Att model
    rfl_model = RandomForestRegressor(**best_params_rfl, random_state=42, oob_score=True, bootstrap=True)
    rfl_model.fit(X_train_synop, y_train_rfl_synop)
    y_pred_rfl = rfl_model.predict(X_test_synop)
    rfl_rmse = np.sqrt(mean_squared_error(y_test_rfl_synop, y_pred_rfl))
    rfl_r2 = r2_score(y_test_rfl_synop, y_pred_rfl)

    # Store metrics
    metrics[f"FSO_Att_{synop_code}"] = {'OOB_Score': fso_model.oob_score_, 'RMSE': fso_rmse, 'R2': fso_r2}
    metrics[f"RFL_Att_{synop_code}"] = {'OOB_Score': rfl_model.oob_score_, 'RMSE': rfl_rmse, 'R2': rfl_r2}

# Step 5: Calculate Average RMSE Across All Models
fso_rmse_list = [m['RMSE'] for k, m in metrics.items() if 'FSO_Att' in k]
rfl_rmse_list = [m['RMSE'] for k, m in metrics.items() if 'RFL_Att' in k]

avg_rmse_fso = np.mean(fso_rmse_list)
avg_rmse_rfl = np.mean(rfl_rmse_list)

# Print Final Metrics
print("\n--- Final Metrics ---")
for key, value in metrics.items():
    print(f"{key}: {value}")

print("\n--- Average RMSE ---")
print(f"FSO_Att Average RMSE: {avg_rmse_fso}")
print(f"RFL_Att Average RMSE: {avg_rmse_rfl}")
```
